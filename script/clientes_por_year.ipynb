```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('../data/clientes_por_ano_csv.csv', header=None, names=['ANIO', 'CLIENTES_ACTIVOS'])
```

```python
print(df.head())
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['ANIO'], df['CLIENTES_ACTIVOS'], marker='o', linewidth=2)
plt.title('Evolución de Clientes de Telefonía')
plt.xlabel('Año')
plt.ylabel('Clientes Activos')
plt.grid(True)
plt.gca().spines[['top', 'right']].set_visible(False)
plt.show()
```
